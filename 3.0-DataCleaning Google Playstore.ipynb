```python
df_copy['Size']=df_copy['Size'].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
df_copy['Size']=df_copy['Size'].replace('Varies with device', np.nan)
```

```python
df['Installs'].unique()
```

```python
df['Price'].unique()
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs', 'Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(item, '')
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype(int)
df_copy['Price']=df_copy['Price'].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Last Updated'].unique()
```

```python
pd.to_datetime(df_copy['Last Updated'])
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
```

```python
numeric_features=[feature for feature in df_copy.columns if df_copy[feature].dtype!='0']
categorical_features=[feature for feature in df_copy.columns if df_copy[feature].dtype=='0']


#print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features: {}'.format(len(categorical_features), categorical_features))
```

```python
## proportion of count data on categorical columns

for col in categorical_features:
    print(df[col].value_counts(normalize=True)*100)
    print('---------------------------')
```

```python
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Numercial Features', fontsize=15, fontweight='bold', alpha=0.9, y=1.)

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df_copy[numeric_features[i]], shade=True, color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

```python
plt.figure(figsize=(20,15))
plt.suptitle('Univariate Analysis Categorical Features', fontsize=15, fontweight='bold', alpha=0.9, y=1.)
category=['Type', 'Content Rating']

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.countplot(x=df_copy[category[i]],palette='Set2')
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```

```python
## Top 10 App Categories

category=pd.DataFrame(df['Category'].value_counts())
category.rename(columns={'Category':'Count'}, inplace=True)
```

```python
category
```

```python
category=category.reset_index()
category.columns=['Category', 'Count']

plt.figure(figsize=(15,6))
sns.barplot(x=category.index[:10], y='Count', data=category[:10], palette='hls')
plt.title('Top 10 App Categories')
plt.xticks(rotation=90)
plt.show()
```

```python
df_copy['Installs'] = df_copy['Installs'].replace('[+,]', '', regex=True).astype(int)

# Group by Category and sum the installs
category_installs = df_copy.groupby('Category')['Installs'].sum().reset_index()

# Sort and get top 10 categories
top_10_categories = category_installs.sort_values(by='Installs', ascending=False).head(10)

# Plot the result
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 6))
sns.barplot(x='Category', y='Installs', data=top_10_categories, palette='hls')
plt.title('Top 10 Categories by Total Installs')
plt.xticks(rotation=90)
plt.show()
```
